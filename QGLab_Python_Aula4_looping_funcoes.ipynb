# Estruturas de Controle

Até agora escrevemos códigos que eram sempre executados linearmente, porém muitas vezes é desejável que partes de um códigos sejam executadas de maneira não linear, seja por meio da **repetição** de instruções um certo número de vezes ou pela execução **condicional** de um trecho de código.

##Condicionais

Este tipo de estutura de controle permite a execução de um trecho de código **se** uma condição for satisfeita.

Para isto, em Python utilizamos a palavra **`if`** seguida de uma expressão e dois pontos, e nas linhas abaixo o código a ser executado.

Geralmente a expressão envolve comparação entre dois ou mais valores, e para isso existem os chamados Operadores de Comparação, mostrados na tabela a seguir

Símbolo | Operação
--- | ---
== | Igualdade
!= | Diferença
> | Maior
>= | Maior ou igual
< | Menor
<= | Menor ou igual

Na célula abaixo, a expressão mostrada (**`n1 < n2`**) trata-se da comparação para verificar se a primeira variável é menor do que a segunda, por meio do operador **`<`**

```
n1 = 4
n2 = 2e4
n1 <= n2
```

```
n1 = 4
n2 = 2e4
n1 != n2
```

```
nome = 'Zé'
sobrenome = 'Silva'

if nome:
  print('Olá ' + nome)
```

```
n1 = 3
n2 = 2

if n1 < n2:
  print('Condição falsa')
if n1 > n2:
  print('Condição verdadeira')
```

A expressão que sucede o **`if`** (**`n1 < n2`** no exemplo) será resolvida e seu resultado determinará se o bloco de código "dentro" do **`if`** será executado. Ao final, o resultado da expressão analisada é um valor Booleano, ou seja, verdadeiro ou falso. Em Python, também podemos fazer uma expressão para o condicional cujo resultado seja numérico, e quando isso ocorre, **apenas o valor 0 é considerado como falso (falsy)** e qualquer **outro valor é considerado verdadeiro (truthy)**. Outros valores falsy são None e String vazia (**`''`**)




```
if 6-7:
  print('Executará quando o resultado for diferente de 0')
  print('Sempre executará independente do resultado')
```

Outro grande uso de estruturas condicionais é a decisão entre dois caminhos que o código pode tomar: **se** a expressão for verdadeira execute um trecho, **senão** execute outro.

Para isso, existe uma outra palavra para sinalizar o outro trecho de código, e esta palavra é **`else`**.

Abaixo temos um exemplo que executa um trecho de código caso o número seja par, e outro trecho caso contrário.

```
num = 3

if num % 2 == 0:
  print('É par!')
else:
  print('É ímpar')
```

```
n1 = 4
n2 = 2e4

if n1==(4) and n2>=(10e5):
  print('OK')
else:
  print('Não OK')
```

Além desses casos, também é possível checar diferentes condições em sequência, e isso pode ser feito de duas maneiras: 1) colocando vários **`if`** em seguida; 2) utilizando **`elif`**.

A primeira solução pode resolver algumas situações, porém traz alguns problemas. Um desses problemas é o uso do **`else`**, pois ele "fecha" o **`if`** mais próximo a ele, e isso pode acarretar comportamentos indesejados, como indicado abaixo.

Nesse exemplo todos os **`if`** sempre serão executados, idependente de uma condição já ter sido verdadeira. Deste modo, mesmo que a primeira condição seja verdadeira, os **`if`** a seguir também serão testados. Assim, o bloco de código referente ao **`else`** será executado sempre que o valor não satisfazer a última condição

```
var = 8

if var == 1:
  print("Igual a 1")
if var == 2:
  print("Igual a 2")
if var <= 5:
  print("Igual ou menor que 5")
else:
  print('Nenhuma das anteriores')
```

No exemplo anterior fizemos uso da operação de comparação (**`==`**), porém caso a comparação envolvesse comparação para ver valores maiores ou menores, mais comportamentos indesejados podem surgir

```
var = 2

if var >= 2:
  print("Maior ou igual a 2")
  if var > 5:
    print("Maior que 5")
if var > 1:
  print("Maior que 1")
else:
  print('Não é maior que 1')
```

Para evitar alguns comportamentos indesejados, podemos fazer o uso de **`else`** seguido de **`if`**. Sua funcionalidade é checar uma outra condição **somente se a anterior falhar**

```
var = 5

if var == 1:
  print("Igual a 1")
else:
  if var == 2:
    print("Igual a 2")
  else:
      if var == 5:
        print("Igual a 5")
      else:
        print("Não é igual a nenhum dos valores")
```

O problema levantado anteriormente foi solucionado, porém ao escrever o código dessa maneira, o código vai ficando cada vez mais deslocado para a direita, o que muitas vezes pode atrapalhar a legibilidade em algumas situações (famigerado **código Hadouken**)

image.png

Para utilizar essa solução e não deslocar muito o código para a direita, existe a palavra-chave **`elif`**, que tem a mesma funcionalidade, porém deixa o código mais legível

```
var = 5

if var == 1:
  print("Igual a 1")
elif var == 2:
  print("Igual a 2")
elif var == 5:
  print("Igual a 5")
else:
  print("Não é igual a nenhum dos valores")
```

```
# Adaptando o exemplo da comparação Maior que
var = 4

if var > 5:
  print("Maior que 5")
elif var > 2:
  print("Maior que 2, menor ou igual a 5")
elif var > 1:
  print("Maior que 1, menor ou igual a 2")
else:
  print("Não é Maior que nenhum dos valores")
```

Apesar de ser uma solução para bastante problemas, o uso demasiado de **`if`** pode atrapalhar a legibilidade de um código, principalmente quando as condições são mais complexas, como no exemplo a seguir

# Definir uma função que verifique se um determinado ano é bissexto usando as condições lógicas and, or, if, elif e/ou else.

```
def ano_bissexto(ano):
  if ano % 400 == 0:
    return True
  elif ano % 4 == 0:
    if ano % 100 != 0:
      return True
  return False

#print(ano_bissexto(1900))
#print(ano_bissexto(ano=int(input())))
#print(ano_bissexto(int(input())))
```

Função usando estrutura condicional

```
def ano_bissexto(ano):
  if ano % 400 == 0 or (ano % 4 == 0 and ano % 100 != 0):
      return True
  return False

#print(ano_bissexto(1600))
print(ano_bissexto(int(input())))
```

## Laços de Repetição (Loopings)

Outra maneira de se controlar o fluxo de um código é fazendo uso dos Laços de Repetição, que são responsáveis pela execução repetida de um bloco de código um certo número de vezes, sendo esse número de vezes definido por meio de uma condição.

Em Python temos dois tipos de Laços de Repetição: **`for`** e **`while`**

### *For*

Esse tipo de laço é geralmente utilizado quando se sabe a princípio quantas vezes um bloco de código precisa ser repetido. A ideia deste laço é executar um bloco de código **para** cada iteração em uma quantidade determinada de iterações.

    para contador em intervalo(0,10):
        imprima(contador)


```
for numeros in range(0,10):
  print(numeros)
```

Na célula acima o número de iterações foi determinado pela função **`range(0,10)`** pois o resultado dessa função produz um objeto **iterável**.

**Iteráveis** são objetos que podem ter seus elementos iterados um a um e, deste modo, servem para uso nos laços **`for`** e algumas outras funções. Como exemplos de objetos iteráveis temos listas, tuplas, entre outros que serão vistos mais adiante.

### *While*

Esse tipo de laço é geralmente utilizado quando não se sabe ao certo quantas iterações serão necessárias. Deste modo esse laço é responsável por executar um bloco repetidamente **enquanto** uma condição for verdadeira.

    enquanto total<100:
        imprima(total)
        total = total * 2

De maneira geral, o que o trecho de pseudocódigo acima faz é verificar se o resultado da expressão (**`total<100`**) é verdadeiro.

1. Caso o resultado seja **falso**, o bloco de código não é executado e a execução continua do trecho de código subsequente ao bloco.
2. Caso o resultado seja **verdadeiro** o bloco de código é executado. Ao final da execução do bloco, a condição é testada de novo para verificar se o bloco será executado novamente. Ao contrário do que ocorre com o **for**, a variável utilizada na expressão não é atualizada automaticamente, precisando deste modo ser atualizada dentro do bloco, como foi feita na primeira linha do bloco (e se não for?)

A célula abaixo mostra a execução de um código Python referente ao pseudocódigo explicado

```
x = 9
while x < 10:
    print(x)
    x += 1
else:
    print("fim while")
```

### Palavra chave *Break*

Muitas vezes é comum precisar sair imediatamente de um laço caso alguma condição seja satisfeita, ignorando o restante das linhas presentes no bloco, ou seja, quando desejamos **quebrar** o laço. Para que possamos fazer isso, existe a palavra chave **`break`**

# Laços de repetição associados a estruturas condicionais

```
# Conversar com um papagaio até enjoar e dizer tchau
print('Conversa com papagaio')

while True:
  resposta = input('Usuário: ')
  print('Papagaio:',resposta)
  if resposta == 'tchau':
    print('Papagaio: Tchau!')
    break
    print('Esta linha de código nunca será executada!')
print('Fim')
```

```
#Encontrar o menor número entre 0 e 100 cujo quadrado seja maior que 800
for i in range(0,100):
  if i**2 > 800:
    print(i)
    break
```

### Palavra chave *Continue*

Outra palavra chave utilizada para controlar um laço é a palavra **`continue`**, quando se deseja que a execução do laço volte para o início, como se ao satisfazer uma condição o laço ignorasse as linhas abaixo e continuasse a iteração a partir do começo do laço.

```
print('Digite seu nome')
nome = input()
sucesso = False

for i in range(0,3):
  print('Olá,',nome,'! Por favor digite a senha')
  senha = input()

  if not senha:
    print('Senha vazia')
    continue

  if senha == '1234':
    print('Senha correta!')
    sucesso = True
    break

if sucesso:
  print('Olá,',nome,'! Senha correta, login efetuado com sucesso!')
  #sucesso = True
else:
  print('Login bloqueado')
```
